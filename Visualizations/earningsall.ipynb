```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/Users/zacdelusignan/Desktop/incomeall.csv')
df['Earnings'] = df['Earnings'].str.replace('$', '\$')
```

```python
custom_palette = ('#2d76be', '#E8C200')

sns.set(style="whitegrid", font="Futura")
g = sns.catplot(
    data=df, kind="bar",
    x="Earnings", y="Percentage", hue="Status",
    ci="sd", palette=custom_palette, alpha=1, height=6
)
g.despine(left=True)
g.set_yticklabels(("0 ", "10 ", "20 ", "30 ", "40 ", "50 ", "60%"))
g.set_axis_labels("", "")
g.legend.set_title("")
g.fig.suptitle("Weekly Earnings",
                y=1.02)

for axes in g.axes.flat:
    axes.set_xticklabels(axes.get_xticklabels(), 
                         rotation=45, 
                         horizontalalignment='right')
```
