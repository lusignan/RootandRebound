```python
#Load libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Load data and make the dollar sign usable
df = pd.read_csv('/Users/zacdelusignan/Desktop/income.csv')
df['Earnings'] = df['Earnings'].str.replace('$', '\$')
```

```python
sns.set(style="darkgrid", font="Gotham")
#Plot
g = sns.catplot(
    data=df, kind="bar",
    x="Earnings", y="Percentage", hue="Status",
    ci="sd", palette="RdBu", alpha=1, height=6
)
g.despine(left=True)
g.set_yticklabels(("0 ", "5 ", "10 ", "15 ", "20 ", "25 ", "30 ", "35 ", "40%"))
g.set_axis_labels("", "")
g.legend.set_title("")
g.fig.suptitle("Weekly Earnings",
                y=1.02)

#Rotate x axis labels
for axes in g.axes.flat:
    axes.set_xticklabels(axes.get_xticklabels(), 
                         rotation=45, 
                         horizontalalignment='right')
```
