```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/zacdelusignan/Desktop/agerange.csv')
```

```python
custom_palette = ('#2D76BE', '#E8C200', '#CB5328', '#00AF54', '#7D3B7B')
sns.set(style='white',font='Futura')
ax = sns.barplot(y='Age', x='Percentage',
data=df,palette=custom_palette)
sns.despine(left=True)
ax.set_xticklabels(('0 ', '5 ', '10 ', '15 ', '20 ', '25 ', '30 ', '35%', '40%'))
ax.set_title("Age Range", size=16)
plt.xlabel("")
plt.ylabel("")
```
